```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(421)
```

```python
# Means
m1 = np.array([0.0, 2.5])
m2 = np.array([-2.5, -2.0])
m3 = np.array([2.5, -2.0])
means = np.array([m1, m2, m3])

"""
FOR DEBUGGING:
print("m1: " + str(m1))
print("m2: " + str(m2))
print("m3: " + str(m3))
"""

print("means:\n" +  str(means))
```

```python
# Covariances
c1 = np.array([[3.2, 0.0], [0.0, 1.2]])
c2 = np.array([[1.2, -0.8], [-0.8, 1.2]])
c3 = np.array([[1.2, 0.8], [0.8, 1.2]])
class_covariances = np.array([c1, c2, c3])

"""
FOR DEBUGGING:
print("c1:\n" + str(c1))
print("\nc2:\n" + str(c2))
print("\nc3:\n" + str(c3))
"""

print("class_covariances:\n" + str(class_covariances))
```

```python
# Sizes
n1 = 120
n2 = 90
n3 = 90
```

```python
# Generating random samples from a multivariate normal distribution
points1 = np.random.multivariate_normal(m1, c1, n1)
points2 = np.random.multivariate_normal(m2, c2, n2)
points3 = np.random.multivariate_normal(m3, c3, n3)

points = np.concatenate((points1, points2, points3))

# Generate corresponding labes
y = np.concatenate((np.repeat(1, n1), np.repeat(2, n2), np.repeat(3, n3)))

plt.figure(figsize = (10, 10))
plt.plot(points1.T[0], points1.T[1], 'ro')
plt.plot(points2.T[0], points2.T[1], 'go')
plt.plot(points3.T[0], points3.T[1], 'bo')

plt.axis('equal')
plt.show()
```

```python
#Get number of classes and number of samples
K = np.max(y)
N = points.shape[0]
```

```python
# Calculate sample_means
sample_means = np.array([np.mean(points[y == (c + 1)], axis=0) for c in range(K)])
print("sample_means:\n" + str(sample_means))
```

```python
# Calculate sample_covariances
sample_covariances = [np.cov(points[y == (c + 1)].T) for c in range(K)]

print("sample_covariances1:\n" + str(sample_covariances[0]))
print("\nsample_covariances2:\n" + str(sample_covariances[1]))
print("\nsample_covariances3:\n" + str(sample_covariances[2]))
```

```python
# Calculate prior probabilities
class_priors = [np.mean(y == (c + 1)) for c in range(K)]
print(class_priors)
```

```python
# Wc, wc, wc0
W = [(-0.5 * np.linalg.inv(sample_covariances[c])) for c in range(K)]
w = [(np.linalg.inv(sample_covariances[c]) @ sample_means[c]) for c in range(K)]
w0 = [(-0.5 * (sample_means[c].T @ sample_covariances[c] @ sample_means[c]) - 0.5 * np.log(np.linalg.det(sample_covariances[c])) + np.log(class_priors[c])) for c in range(K)]

# Needed to convert to arrays because cannot take the transpose of a list
W = np.array(W)
w = np.array(w)
```

```python
# gc(x) function
def g(points):
    predictions = []
    for i in range(N):
        results = [points[i].T @ W[c] @ points[i] + w[c].T @ points[i] + w0[c] for c in range(K)]
        predictions.append(results)
    return predictions
```

```python
y_predictions = g(points)
y_pred = np.argmax(y_predictions, axis = 1) + 1

# Confusion matrix
confusion_matrix = pd.crosstab(y_pred, y, rownames = ['y_predicted'], colnames = ['y_truth'])
print(confusion_matrix)
```

```python
# Visualization
# evaluate discriminant function on a grid
x1_interval = np.linspace(-6, +6, N)
x2_interval = np.linspace(-6, +6, N)

x1_grid, x2_grid = np.meshgrid(x1_interval, x2_interval)
discriminant_values = np.zeros((len(x1_interval), len(x2_interval), K))
discriminant_values_f = np.zeros((len(x1_interval), len(x2_interval), K))

for c in range(K):
    discriminant_values[:,:,c] = W[c, 0, 0] * pow(x1_grid, 2) + W[c, 0, 1] * x1_grid * x2_grid + W[c, 1, 0] * x1_grid * x2_grid + w[c, 0] * x1_grid + W[c, 1, 1] * pow(x2_grid, 2) + w[c, 1] * x2_grid + w0[c]

# Same values will be used for coloring of the areas
discriminant_values_f[:,:,0] = discriminant_values[:,:,0]
discriminant_values_f[:,:,1] = discriminant_values[:,:,1]
discriminant_values_f[:,:,2] = discriminant_values[:,:,2]


# For contour
A = discriminant_values[:,:,0]
B = discriminant_values[:,:,1]
C = discriminant_values[:,:,2]

# For contourf
A_f = discriminant_values_f[:,:,0]
B_f = discriminant_values_f[:,:,1]
C_f = discriminant_values_f[:,:,2]


# For contour
A[(A < B) & (A < C)] = np.nan
B[(B < A) & (B < C)] = np.nan
C[(C < A) & (C < B)] = np.nan

# For contourf
A_f[(A_f > B_f) & (A_f > C_f)] = np.nan
B_f[(B_f > A_f) & (B_f > C_f)] = np.nan
C_f[(C_f > A_f) & (C_f > B_f)] = np.nan


# For contour
discriminant_values[:,:,0] = A
discriminant_values[:,:,1] = B
discriminant_values[:,:,2] = C

# For contourf
discriminant_values_f[:,:,0] = A_f
discriminant_values_f[:,:,1] = B_f
discriminant_values_f[:,:,2] = C_f



plt.figure(figsize = (10, 10))
plt.plot(points[y == 1, 0], points[y == 1, 1], "r.", markersize = 10)
plt.plot(points[y == 2, 0], points[y == 2, 1], "g.", markersize = 10)
plt.plot(points[y == 3, 0], points[y == 3, 1], "b.", markersize = 10)
plt.plot(points[y_pred != y, 0], points[y_pred != y, 1], "ko", markersize = 12, fillstyle = "none")

plt.contour(x1_grid, x2_grid, discriminant_values[:,:,0] - discriminant_values[:,:,1], levels = 0, colors = "k")
plt.contour(x1_grid, x2_grid, discriminant_values[:,:,0] - discriminant_values[:,:,2], levels = 0, colors = "k")
plt.contour(x1_grid, x2_grid, discriminant_values[:,:,1] - discriminant_values[:,:,2], levels = 0, colors = "k")

plt.contourf(x1_grid, x2_grid, discriminant_values_f[:,:,1] - discriminant_values_f[:,:,2], levels = 0, colors = "r", alpha=.25)
plt.contourf(x1_grid, x2_grid, discriminant_values_f[:,:,0] - discriminant_values_f[:,:,2], levels = 0, colors = "g", alpha=.25)
plt.contourf(x1_grid, x2_grid, discriminant_values_f[:,:,0] - discriminant_values_f[:,:,1], levels = 0, colors = "b", alpha=.25)

plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```
